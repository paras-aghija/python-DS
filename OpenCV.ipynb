# OpenCV

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('naruto.jpg')
# opencv by default reads an image in BGR mode
```

```python
plt.imshow(img)
plt.show()
```

```python
newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(newImg)
plt.show()
```
