```python
top_10 = df.groupby('Country/Region').sum().sort_values(by='Confirmed', ascending=False).head(10)[['Confirmed', 'Deaths','Recovered']]
print(top_10)
```

Turkey has lowest number of confirmed cases in the top_10 dataframe

```python
top_10['Recovered_percentage'] = (top_10['Recovered']/ top_10['Confirmed'])*100
top_10['Death_percentage'] = (top_10['Deaths']/ top_10['Confirmed'])*100
```
