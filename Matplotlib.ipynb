### Line Plots

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
plt.plot(x, y1, color="red", label="Apple", linestyle='dashed')
plt.plot(x, y2, color="green", label="Kiwi", marker='o')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruits over time")
plt.legend()
plt.show()
```

### Scatter Plots

```python
# adjust size of plot
plt.figure(figsize=(20,10))
plt.scatter(x, y1, color="red", label="Apple", marker='^')
plt.scatter(x, y2, color="green", label="Kiwi", marker='o')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot - Prices of fruits over time")
plt.legend()
plt.show()
```

### Bar Graphs

```python
plt.bar?
```

```python
plt.style.use('dark_background')
x_cord = np.array([0,1,2])*2
plt.bar(x_cord, [10,20,15], width=0.5, label="Current Year", tick_label=['Gold', 'Platinum', 'Silver']) #current year
plt.bar(x_cord+0.5, [20,10,12], width=0.5, label="Next Year", color="orange") # next year
plt.title("Metals price comparison")
plt.xlabel('Metal')
plt.ylabel('Price')
plt.ylim(0,30)
plt.xlim(-1,5)
plt.legend()
plt.show()
```

### Pie Chart

```python
subjects = ["maths", "chemistry", "physics", "english"]
weightage = [20,10,15,5]
plt.style.use("dark_background")
plt.pie(weightage, labels=subjects, explode=(0,0,0.1,0), autopct='%1.1f%%', shadow=True)
plt.show()
```

### Movie data visualisation

```python
df = pd.read_csv("https://raw.githubusercontent.com/paras-aghija/python-DS/main/datasets/movies.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
titles = list(df.get('Title'))
titles[:5]
```

```python
freq_title = dict()
for title in titles:
    length = len(title)
    if(freq_title.get(length)) is None:
        freq_title[length] = 1
    else:
        freq_title[length] += 1

print(freq_title)
```

```python
X = np.array(list(freq_title.keys()))
Y = np.array(list(freq_title.values()))
print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of movie title")
plt.ylabel("no. of movies")
plt.title("Movie Data visualisation")
plt.show()
```

### Normal Distribution

### Multivariate Normal Distribution

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0],[0,1]])

dist = np.random.multivariate_normal(mean, cov, 500)

plt.scatter(dist[:,0], dist[:,1])
plt.show()
```

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],[0.8,1]])

dist = np.random.multivariate_normal(mean, cov, 500)

plt.scatter(dist[:,0], dist[:,1])
plt.show()
```
